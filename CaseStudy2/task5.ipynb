```python
import numpy as np
import pandas as pd
import seaborn as sns
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.groupby(["time"])["total_bill"].agg(["sum", "min", "max", "mean"])
```

```python
df.groupby(["day","time"])["total_bill"].agg(["sum", "min", "max", "mean"])
```

```python
lunch_female = df[(df["time"] == "Lunch") & (df["sex"] == "Female")]
lunch_female.groupby("day")[["total_bill", "tip"]].agg(["sum", "min", "max", "mean"])
```

```python
df.loc[(df["size"] < 3) & (df["total_bill"] > 10), "total_bill"].mean()
```

```python
df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
```

```python
df.head()
```

```python
df.sort_values("total_bill_tip_sum", ascending=False, inplace=True)
```
